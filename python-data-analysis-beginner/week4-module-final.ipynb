## 3. 예외 처리 (Exception Handling)

```python
# 강제로 어떠한 숫자를 '0'으로 나누어봅시다. 프로그램이 비정상적으로 종료될 것입니다.
print 10/0

# 이러한 예외상황(Exception)에서 프로그램을 종료시키지 않고 프로그래머가 이를 제어할 수 있는 방법이 있습니다.

try:
    print "0으로 나누어볼까요?"
    print 10/0
    print "짜잔! 성공했습니다!"
except:
    print "실패했습니다! 당연히 0으로 나눌 수는 없겠지요."
    
    
# 이를 디버깅을 해보면, try문 안을 실행하다가 예외가 발생하는 순간 바로 except 문으로 빠지는 것을 알 수 있습니다.

# 정상적인 구문이라면 except는 실행되지 않을 것입니다.

try:
    print "Hello"
except:
    print "Bye"

    
# 어떠한 종류의 예외가 발생했는지 알아볼 수도 있습니다.

try:
    print 10/0
    
except Exception as e:  
    print e

try :
    print 10/0
except ZeroDivisionError as e :
    print 'cannot divide by 0'
    print e

    
# 생경한 문법이죠?
# 예외가 발생하면, except문에서는 이 예외를 오른쪽의 예외 객체에 담아준답니다.
# 예외는 여러가지 종류가 있기 때문에 다양한 종류의 예외 클래스들이 존재하는데, 그 예외 클래스들의 가장 부모 클래스는 바로 이 'Exception' 클래스랍니다.
# 모든 예외들의 기본 형이고, 다른 예외들은 모두 이 클래스에서 상속받은 클래스라고 할 수 있겠지요.
# 어째뜬 저 문법은 '지금 발생한 예외를 Exception 클래스로 받아볼꺼고, 이걸 e라는 변수에 할당하겠어'라는 뜻입니다.
# 실제로 해당 예외의 에러메시지인 'integer division or modulo by zero'가 나타나는 것을 알 수 있습니다.
# e는 예외 객체이기 때문에, 실제 디버깅해보면 에러 메시지 뿐만 아니라 다양한 정보를 가지고 있다는 것을 알 수 있습니다.

# 위의 구문을 try-except (혹은 try-catch) 구조라고 한답니다.
# 이러한 것을 통칭 '예외처리 하다'라고 합니다.

# try-except 외에 try-except-finally 구조도 존재합니다.
# finally문 안의 코드들은 위에서 결과가 어찌되었든 '무조건' 실행됩니다.

try:
    print "숫자형과 문자열형을 더해보겠습니다."
    print "10" + 5
    print "잘 더해졌습니다!"
except Exception as e:
    print "예외가 발생했습니다."
    print "예외의 메시지는 다음과 같습니다: {0}".format(e)
finally:
    print "어쨌든 프로그램의 끝!"
    
    
try:
    print "숫자형과 숫자형을 더해보겠습니다."
    print 10 + 5
    print "잘 더해졌습니다!"
except Exception as e:
    print "예외가 발생했습니다."
    print "예외의 메시지는 다음과 같습니다: {0}".format(e)
finally:
    print "어쨌든 프로그램의 끝!"    


# finally 구문 안에 있는 코드는 예외가 발생하든 안하든 무조건 실행된다는 것을 알 수 있습니다.
# 만약 open()을 이용해서 파일을 열었을 때, 안에서 예외가 발생하더라도(혹은 발생하지 않더라도) 
# 무조건 마지막에 실행되어야 하는 f.close()함수 등을
# 이렇게 finally 안에 넣으면 되겠지요?

try:
    f = open('./try_exception.txt', 'r')
except IOError as IOe:
    print 'no file or directory, make a new one'
    f = open('./try_exception.txt', 'w')
except Exception as e:
    print e
finally:
    f.close()
    
```

### (1) Series

Pandas에는 크게 두가지의 Data Struncture(데이터를 저장하는 구조)가 있습니다. 1차원 데이터는 Series, 2차원 데이터는 DataFrame이라는 구조로 처리를 하고 있는데요, 쉽게 얘기해서 Series는 우리가 그동안 배웠던 리스트이고, DataFrame은 테이블형 데이터구조라고 생각하시면 됩니다.
오늘은 Series를 생성하고 데이터를 다루는 방법을 익히도록 하겠습니다.

```python
import pandas as pd # pandas 패키지를 사용하기 위해서 import 합니다.

# Series는 1차원 데이터를 저장하는 구조입니다.

# 1) 리스트 

# 가장 기본적인 형태는 리스트 형태를 Series로 생성하는 것입니다.
data = pd.Series([1,2,3,4,5])

# 지금부터 우리가 다루는 것은 '의미있는 데이터'들이라고 하면, 데이터에 이름을 붙여주는 게 좋겠죠?
data = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'])


# 이런 식도 가능합니다.
data = [1, 2, 3]
axis = ['x', 'y', 'z']

data = pd.Series(data, index = axis)


# 주의! 리스트 두 개로 index:value 쌍을 만드는 경우에는 데이터와 인덱스는 반드시 길이가 같아야 합니다.
data = [1,2,3]
axis = ['a', 'b', 'c', 'd']
# data = pd.Series(data, index = axis)



##### 길이가 같은 리스트를 한번에 묶어주는 데이터형이 있었던 것 같지 않나요

# 카페이름과 아메리카노의 가격을 따로 저장했던 두개의 리스트를
cafe = ['Starbucks', 'Coffee Bean', 'Twosome Place', 'EDIYA Coffee', 'Cafe Bene']
price = [3800, 4500, 4100, 2900, 4500]

# 딕셔너리의 key:value 쌍으로 만들어줬었죠!
cafe = {'Starbucks' : 3800,
        'Coffee Bean' : 4500,
        'Twosome Place' : 4100, 
        'EDIYA Coffee' : 2900,
        'Cafe Bene' : 4500
        }


# 2) 딕셔너리

# Series의 인덱스와 데이터를 딕셔너리로 묶어줄 수 있겠죠?
americanos = pd.Series(cafe)

# 딕셔너리로 Series를 만드는 경우에는 index를 따로 명시하여 순서를 정하거나 선택을 할 수 있습니다.
americanos = pd.Series(cafe, index = ['Cafe Bene', 'EDIYA Coffee', 'Starbucks'])

# 딕셔너리에 없는 'key'가 들어갔다면?
americanos = pd.Series(cafe, index = ['Cafe Bene', 'EDIYA Coffee', 'Pascucci', 'Starbucks'])


# index 이름과 숫자로 모두 접근 가능합니다.
print americanos['Starbucks']
print americanos[0]

# Slicing
print americanos['Cafe Bene' : 'Pascucci']
print americanos[0:3]
```
