```python
# Dictionary 모양
# {
#   'key1' : 'value1',
#   'key2' : 'value2',
#        ...
# }

# Key는 바꿀 수 없습니다(immutable)
# List는 Index로 순서를 매겼던 것과 달리
# Dictionary는 순서가 없습니다(Unordered set of key:value pairs)

tv_channel = { 'EBS' : 4, 'KBS' : 7, 'MBC' : 10, 'SBS' : 6, 'tvN' : 20 }
print tv_channel
print tv_channel.items()


print tv_channel.keys()

# .keys()의 결과값은 iterate 기능은 가능하지만 
# append, insert, remove 등 리스트 고유한 함수를 사용할 수는 없다.

# 리스트로 만들기
list(tv_channel.keys())

print tv_channel.values()


# 새로운 key:value 쌍 더하기
tv_channel['JTBC'] = 15 


# key로 해당 value에 접근할 수 있습니다.
print tv_channel['EBS'] 
print tv_channel.get('EBS')


tv_channel['EBS'] = 5 # key는 바꿀 수 없지만 value는 바꿀 수 있습니다
print tv_channel


# key가 있는지 없는지
print 'JTBC' in tv_channel
print tv_channel.has_key('EBS')


# key 삭제하기
del tv_channel['EBS']

tv_channel.clear()
```

# 2. 함수

함수는 입력된 데이터에 대해 사용자가 정의한 기능을 수행하여 결과를 반환하는 역할을 합니다.

믹서기를 예로 들어볼까요? 믹서기는 넣은 음식을 갈아주는 역할을 합니다. 그러나 가는 역할은 똑같지만 들어가는 음식에 따라 결과가 달라지죠.
함수를 믹서기에 대입해서 비교해보면, 믹서라는 함수는 가는 역할을 하지만, 들어가는 음식(입력, Input)에 따라서 나오는 결과(반환값, Output)이 달라지게 됩니다.

```python
# 함수 기본 구조
# def 함수이름(입력) :
    # 기능
    # return 반환값
    
# 함수 정의(Definition) 예시
# count_a라는 함수는 주어진 문자열에서 a의 개수를 반환하는 동작을 합니다.
def count_a(string) :
    return string.count('a')


# 함수의 정의 부분은 '이 이름의 함수는 이런이런 입력을 받아서 이런 결과를 반환하는 기능을 하는 함수야'라고
# 컴퓨터에게 알려주는 역할을 하는 것입니다. 실제 이 함수를 사용하기 위해서는 믹서기에 과일을 넣듯 실제 입력값을 넣어줘야 합니다.

# 함수 사용: 호출(Call)
# 위에서 add 함수를 정의할 때 1개의 입력이 필요한 함수로 만들었기 때문에
# 사용할 때도 반드시 입력을 1개 넣어줘야 합니다.
print count_a('apple')

# 상수를 그대로 입력값에 넣을 수도 있지만, 대부분의 경우 이렇게 변수를 통해 값을 넣어줍니다.
string = 'abandon'
print count_a(string)

# 함수의 결과값도 변수에 대입하여 저장할 수 있습니다.
result = count_a(string)
print result



# a를 찾고 싶을 때도 있지만 다른 알파벳을 찾고 싶을 때도 있잖아요?
# 함수의 입력은 여러개를 넣을 수도 있습니다.

def count_alphabet(string, find) :
    return string.count(find)

# count_alphabet 함수는 정의할 때 입력을 2개 달라고 했기 때문에
# 호출할 때에도 반드시 입력 값이 2개 있어야 합니다.

count_alphabet('apple', 'p')
count_alphbet(string, 'a')
result = count_alphbet(string, 'o')


# 일반적인 함수의 형태를 알아보았습니다. 하지만 모든 함수에 입력과 반환값이 있어야 하는 것은 아닙니다.
# 입력 또는 반환값이 없는 함수 형태도 있을 수 있습니다.

# 입력이 없는 함수 형태 예시
def hello():
    return "Hello"

# 함수 호출
print hello()


# 반환값/리턴값이 없는 함수 형태 예시
def add(a, b):
    print a, "와 ", b, ' 의 합은 ', a+b
    
# 함수 호출
add(3, 5)


# 입력과 리턴값이 모두 없는 함수
def print_hello():
    print "Hello"
    
# 함수 호출
print_hello()
```

함수에서 입력으로 받는 값을 Argument라고 하는데요, Argument에 대해서 좀 더 자세히 알아봅시다.

```python
# Default Argument Values

# 앞서 함수의 정의를 설명할 때 말씀드린 것처럼, 함수를 만들 때 정의해 놓은 필요한 입력의 개수만큼 함수를 호출할 때 입력해주어야 합니다.
# 예를 들어, 
def add(a, b):
    return a+b
# 와 같이 a와 b 두개의 입력이 필요한 함수를 정의했다면, add 함수를 호출할 때에도
print add(3, 4) 
# 이처럼 입력을 반드시 2개 넣어주어야 합니다.

# 두 개 이상의 입력값을 받아야 하는 함수의 경우, 일부 입력값(Argument)에 기본 설정값을 정해두면,
# 사용자가 굳이 매번 값을 입력해주지 않아도 함수를 사용할 수 있게 됩니다. 이를 Default Argument라고 합니다.

# alphabet 세기를 다시 해볼까요?
# 특정한 알파벳을 원하면 두번째 argument에 입력하면 되지만, 그렇지 않은 경우 'a'를 찾게 합니다.

# Default Argument는 항상 Non-default Argument 뒤에 나열되어야 합니다.
# Non-default Arguments들이 모두 나열되고 그 뒤에 Default Arguments를 나열해야 하는거죠.

def count_alphabet(string, find = 'a') :
    return string.count(find)

count_alphabet('apple', 'p')
count_alphabet('apple')
```

```python
# 함수의 결과값은 항상 하나여야 합니다.

# 입력값 두 개를 받아서 몫과 나머지를 리턴하는 함수를 만들고 싶습니다
def div_and_mod(a, b) :
    return a / b, a % b

# 이렇게 작성하면 결과가 어떻게 나올까요

result = div_and_mod(10, 3)
print result

# return 뒤의 a/b, a%b가 Tuple packing에 의해 (a/b, a%b) 튜플 하나가 리턴됩니다.

div, mod = div_and_mod(10, 3) # Sequence Unpacking 성질을 이용할 수 있겠죠?
print div, mod 


# 그렇다면 이런 모양은 어떻게 될까요
def div_and_mod(a, b) :
    return a / b
    return a % b

print div_and_mod(10, 3)

# 함수의 정의에서 return 을 만나면 함수는 그 줄을 마지막으로 수행이 종료됩니다.
```
